## 1.2 Считывание датасета

```python
# функция для преобразования дат (int -> datetime)
def get_time(time):
    if type(int(str(time)))==int:
        seconds = (int(time) - 25569) * 86400
        temp = datetime.utcfromtimestamp(seconds)
        return temp
        
    elif type(time.astype('int')) == np.int32 or numpy.int64:
        seconds = (int(time) - 25569) * 86400
        temp = datetime.utcfromtimestamp(seconds)
        return temp
    
    else:
        return np.nan

vget_time = np.vectorize(get_time)

# функция для поиска количества повторений
def count_freq(df: pd.DataFrame, type_to_count: 'str') -> pd.DataFrame:

    freq_df = pd.DataFrame(
        columns=[f'{type_to_count.upper()}_freq_type_{ttype}' for ttype in set(df[type_to_count])],
        index=pd.Index(set(df['tcs_customer_id']))
    ).fillna(0)

    temp_df = df[['tcs_customer_id', type_to_count]]
    for _, tcs_id in enumerate(tqdm_notebook(set(temp_df['tcs_customer_id']))):

        count_val = temp_df[temp_df['tcs_customer_id'] == tcs_id][type_to_count]
        dict_val = dict(Counter(count_val))

        for key in dict_val.keys():
            freq_df.loc[tcs_id, f'{type_to_count.upper()}_freq_type_{key}'] = dict_val.get(key)
                               
    return freq_df
```

```python
# читаем датасет
df_read = pd.read_csv('SAMPLE_ACCOUNTS.csv', sep=';', decimal=',',
                dtype={'pmt_freq': str, 'fact_close_date':str})


id_accounts = pd.read_csv('SAMPLE_CUSTOMERS.csv', sep=';')
id_accounts.rename(columns={'bad':'default'}, inplace=True)

print('Размер датасета:', df_read.shape)
print('Количество заемщиков:', id_accounts.shape[0])
```

```python
# у нас есть 2 таблицы - сам датасет и указание на train/test.
# объединим их в одну таблицу и удалим данные с пометкой test (потому что отсутствует таргет)

df = df_read.merge(id_accounts, how='left', on='tcs_customer_id')
df = df[df.sample_type == 'train'].drop('sample_type', axis=1).reset_index(drop=True)
df['default'] = df['default'].astype(int)

print(df.shape)
df.head()
```

Частота выплат

```python
plt.figure(figsize=(15, 15))
plt.subplots_adjust(wspace=0.3, hspace=0.5)

cols = ['ttl_delq_5', 'ttl_delq_5_29', 'ttl_delq_30', 'ttl_delq_30_59', 
        'ttl_delq_60_89', 'ttl_delq_90_plus']


ncols = 2
nrows = len(cols) // ncols + (len(cols) % ncols > 0)

for ind, col in enumerate(cols):
    
    temp = df.query(f'0 < {col} <= 5')
    ax = plt.subplot(nrows, ncols, ind + 1)
    sns.distplot(temp[col], bins=110, color = (np.random.random(), np.random.random(), np.random.random()))
    plt.title(f'{col} distplot')

plt.show()


# вывод:

# расшифровка:
# TTL_DELQ_5        Количество просрочек до 5 дней
# TTL_DELQ_5_29     Количество просрочек от 5 до 29 дней
# TTL_DELQ_30_59    Количество просрочек от 30 до 59 дней
# TTL_DELQ_60_89    Количество просрочек от 60 до 89 дней
# TTL_DELQ_30       Количество просрочек до 30 дней
# TTL_DELQ_90_PLUS  Количество просрочек 90+ дней



# в целом,
# заемщики чаще всего не выплачивают 1 раз в разные периоды
```

```python
tlls = [
 'ttl_delq_5',
 'ttl_delq_5_29',
 'ttl_delq_30_59',
 'ttl_delq_60_89',
 'ttl_delq_30',
 'ttl_delq_90_plus'
]
```

```python
# создадим таблицу, которая отображает, насколько 1 просроченный платеж влияет на статус дефолта
# индекс - количество, сколько раз заемщик пропустил выплату
# колонки - период

# пример:
# ttl_delq_5 - Количество просрочек до 5 дней.
# заемщики, которые допустили просрочку 1 раз в период ttl_delq_5,
# дефолтятся в 9.995% (0.099952 * 100%, почти 10%) случаев




tll_data = pd.DataFrame()
for tll in tlls:
    df_ttl = pd.DataFrame(df[df[tll]<10].groupby([tll, 'default'])[tll].count())
    df_ttl.rename(columns={tll:'ttl'}, inplace=True)
    df_ttl['%'] = df_ttl['ttl'] / df_ttl.groupby(tll)['ttl'].transform('sum')

    df_ttl.reset_index(drop=False, inplace=True)
    df_ttl.drop('ttl', axis=1, inplace=True)
    tll_data[tll] = df_ttl[df_ttl['default']==1]['%']
    
tll_data.reset_index(drop=True, inplace=True)
```

```python
tll_data
```

```python
# посмотрим, сколько нужно допустить просрочек по выплате в каждый период, 
# чтобы достичь максимальной вероятности дефолта

for col in tll_data:
    print(col)
    print('max % default 1:  ', round(np.max(tll_data[col]), 4), '||  count =', np.argmax(tll_data[col]), '\n')
```

```python
# Аналогично для минмальной вероятности дефолта

for col in tll_data:
    print(col)
    print('max % default 1:  ', round(np.min(tll_data[col]), 4), '||  count=', np.argmin(tll_data[col]), '\n')
```
