```python
def do_smooth(df, n_window):
    df['Open_not_smooth'] = df['Open']
    df['Open'] = round(df['Open'].rolling(window=n_window).mean().iloc[n_window-1:], 2)
    df = df[n_window:]
    return df


def download_stocks(ticker = 'GAZP', start = '2020-01-01', end = '2023-01-01', interval=24):
 time.sleep(0.01) 
 with requests.Session() as session:
    data = apimoex.get_market_candles(session, ticker, start=start, end=end, interval=interval)
    df = pd.DataFrame(data)
    df.columns = ['Date', 'Open', 'Close', 'High', 'Low', 'Value']
    df.set_index('Date', inplace=True)
    df = df[['Open', 'High', 'Low', 'Close', 'Value']]
    # display(pd.concat([df.head(3), df.tail(3)]))
    return df
```

# Monte-Carlo

```python
def roll_dice():
    die_1 = random.randint(1, 6)
    die_2 = random.randint(1, 6)

    # Определим является ли значение на костях одинаковым
    if die_1 == die_2:
        same_num = True
    else:
        same_num = False
    return same_num
```

```python
num_simulations = 1000
max_num_rolls = 1000
bet = 1

# Отслеживаемые переменные
win_probability = []
end_balance = []
```

```python
fig = plt.figure()
plt.title("Monte Carlo Dice Game [" + str(num_simulations) + "  simulations]")
plt.xlabel("Roll Number")
plt.ylabel("Balance [$]")
plt.xlim([0, max_num_rolls])

for i in range(num_simulations):
    balance = [1000]
    num_rolls = [0]
    num_wins = 0
    # Выполняется до тех пор пока игрок не выкинет 1000 раз
    while num_rolls[-1] < max_num_rolls:
        same = roll_dice()
        # Результат если кости одинаковые
        if same:
            balance.append(balance[-1] + 4 * bet)
            num_wins += 1
        # Результат если кости разные
        else:
            balance.append(balance[-1] - bet)

        num_rolls.append(num_rolls[-1] + 1)
    # Сохраняем отслеживаемую переменную и добавляем строку к рисунку
    win_probability.append(num_wins/num_rolls[-1])
    end_balance.append(balance[-1])
    plt.plot(num_rolls, balance)

# Выведем график после завершения моделирования
plt.show()

# Усредненная вероятность выигрыша и конечного баланса
overall_win_probability = sum(win_probability)/len(win_probability)
overall_end_balance = sum(end_balance)/len(end_balance)
# Вывод средних значений
print("Average win probability after " + str(num_simulations) + " runs: " + str(overall_win_probability))
print("Average ending balance after " + str(num_simulations) + " runs: $" + str(overall_end_balance))
```

# Тест роста за последующий промежуток

```python
wins = [3, 5, 7, 10, 14, 17, 22, 25] # 
ru_ticks = [] # 'ROSN', 'GAZP', 'NVTK', 'SBER', 'MGNT', 'PLZL','NVTK', 'SBER', 'MGNT', 'PLZL', 'LKOH', 'GMKN', 'ALRS', 'MGNT', 'POLY', 'YNDX', 'NLMK' 
en_ticks = ['ROSN.ME', 'GAZP.ME', 'NVTK.ME', 'SBER.ME',] # 'MGNT.ME', 'PLZL.ME', 'LKOH.ME', 'GMKN.ME', 
            # 'ALRS.ME', 'POLY.ME', 'YNDX.ME', 'NLMK.ME', 'TSLA', 'NFLX', 'GE', 'AAPL', 'AMZN', 

ticks = ru_ticks + en_ticks

res = pd.DataFrame(index = wins)
av_buy_res = pd.DataFrame(index = wins)
av_sell_res = pd.DataFrame(index = wins)

falls = pd.DataFrame(index = wins)

for names in tqdm(ticks):
  res_lst = []
  av_buy = []
  av_sell = []

  res_fall = []

  if names in en_ticks:
    df = yf.download(names, start='1970-01-01', end='2024-01-01')
  else:  
    df = download_stocks(ticker = names, start = '2015-01-01', end = '2023-06-01', interval=24)

  data = df.filter(['Close'])
  data_frame = data
  data = data.values
  train_len = round(len(data)*0.9)
  sc = MinMaxScaler(feature_range = (0, 1))
  train_scaled = sc.fit_transform(data)
  train_scaled = data

  train_data = train_scaled[0:train_len, :]

  for win in wins:
    x_train = []
    y_train = []
    window = win

    for i in range(window, len(train_data)):
      x_train.append(train_data[i-window:i, 0])
      y_train.append(train_data[i, 0])
    x_train, y_train = np.array(x_train), np.array(y_train)
    x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
    
    ans = []
    ans_buy = []
    ans_sell = []

    ans_fallRise = []

    for i in range(len(x_train)-1):
      if x_train[i][-1] - x_train[i][0] > 0:
        if (x_train[i+1][-1] - x_train[i+1][0] > 0) and x_train[i+1][0][0] != 0:
              ans.append('buy')
              ans_buy.append(np.round(x_train[i+1][-1]/x_train[i+1][0], 3))

        elif (x_train[i+1][-1] - x_train[i+1][0] < 0) and x_train[i+1][0][0] != 0: 
          ans.append('sell')
          ans_sell.append(np.round(1-x_train[i+1][-1]/x_train[i+1][0], 3))
      
      elif x_train[i][-1] - x_train[i][0] < 0:
        if (x_train[i+1][-1] - x_train[i+1][0] < 0) and x_train[i+1][0][0] != 0:
              ans_fallRise.append('fall')

        elif (x_train[i+1][-1] - x_train[i+1][0] > 0) and x_train[i+1][0][0] != 0: 
          ans_fallRise.append('rise')
    
    res_lst.append(round(Counter(ans)['buy']/(Counter(ans)['sell']+Counter(ans)['buy']), 3))
    av_buy.append(np.mean(ans_buy))
    av_sell.append(np.mean(ans_sell))

    res_fall.append(round(Counter(ans_fallRise)['fall']/(Counter(ans_fallRise)['fall']+Counter(ans_fallRise)['rise']), 3))

  res[f'{names}'] = res_lst
  av_buy_res[f'{names}'] = av_buy
  av_sell_res[f'{names}'] = av_sell

  falls[f'{names}'] = res_fall
```

```python
for table in [res, av_buy_res, av_sell_res]: #falls
  # plt.figure(figsize=(len(ticks), round(len(ticks)/2)+3))
  # sns.heatmap(table, annot=True)
  display(table)
  # plt.show()
  print()
```

```python
lst_to_trade = []

for i in wins:
  for j in ticks:
    lst_to_trade.append((j, i))

trade = pd.DataFrame(np.zeros( (len(wins), len(ticks)) ), index=wins, columns=ticks)

pbar = tqdm(total=len(lst_to_trade))

for ticker, window in lst_to_trade:

  lst_money = []

  for i in range(1000):
    money = 100
    for i in range(5):
      choice = random.choices(['buy', 'sell'], weights = [res[ticker][window], 1 - res[ticker][window]])[0]
      if choice == 'buy':
        money = money * av_buy_res[ticker][window]*(1 - 0.0035)
      elif choice == 'sell': 
        money = money * (1 - av_sell_res[ticker][window])*(1 - 0.0035)

    lst_money.append(money)
  
  trade.loc[window, ticker] = np.round(np.mean(lst_money)/money/12, 5)*100
  pbar.update(round(1/len(lst_to_trade)*100))
```

```python
data = df

df.tail()
```

```python
df= data.reset_index()['Open']

scaler = MinMaxScaler(feature_range = (0,1))
df = scaler.fit_transform(np.array(df).reshape(-1,1))
training_size = int( len(df) * 0.75)
testing_size  = len(df) - training_size
train_data , test_data = df[0 : training_size, :], df[training_size : len(df), :1]

def create_dataset(dataset, time_step=1):
    train_X, train_Y = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), 0]
        train_X.append(a)
        train_Y.append(dataset[i + time_step, 0])
    return np.array(train_X), np.array(train_Y)
```

```python
time_step = 50
x_train, y_train = create_dataset(train_data, time_step)
x_test, y_test = create_dataset(test_data, time_step)
```

```python
x_train = x_train.reshape(x_train.shape[0], x_train.shape[1], 1)
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1], 1)
```

```python
model = Sequential()

model.add(Dense(units = 128, input_shape = (time_step, 1), activation='relu'))
model.add(LSTM(units = 128, activation='tanh', return_sequences = True))
model.add(Dropout(0.3))
model.add(LSTM(units = 128, activation='tanh', return_sequences = True))
model.add(Dropout(0.3))
model.add(LSTM(units = 128, activation='tanh', return_sequences = False))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(1, activation='relu'))
```
